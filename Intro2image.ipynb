```python
# RGB to Grayscale Conversion Function
# A simple function named rgb2gray that converts an input RGB image to a grayscale image.
# It calculates the grayscale image by taking a weighted average of the red, green, and blue values for each pixel.
# The weights used are 0.299 for red, 0.587 for green, and 0.114 for blue.
# The result is returned as an integer grayscale image.
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114]).astype(int)
```
